```python
y = df["price"]
x = df[["carat","cut","color","clarity","depth","table","price","x","y","z"]]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=.8,random_state=42)
```

```python
x_train
```

```python
x_train.dtypes
```

```python
x_train_cat = x_train.select_dtypes(include=['object'])
x_train_cat
```

```python
x_train_num = x_train.select_dtypes(include=["float64"])
x_train_num
```

```python
# Here we have inconsistencies and large difference between carret value and depth and table 
# we will apply standardization on this 
from sklearn.preprocessing import StandardScaler
standard_scaler_object = StandardScaler()
```

```python
# ALl the values lies between -3 and +3 
x_train_num_rescaled = pd.DataFrame(standard_scaler_object.fit_transform(x_train_num),columns=x_train_num.columns,index=x_train_num.index)
x_train_num_rescaled
```

```python
 x_train_num.describe()
```

# Calculation of each on own

```python
print("no of numerical features :",standard_scaler_object.n_features_in_)
print("Mean of each columns :",standard_scaler_object.mean_)
print("Standard scaler deviation of each column :",np.sqrt(standard_scaler_object.var_))
```

## Data Preprocessing of categorical values using label encoding

```python
# Here normalize=True gives the percentage of each values
x_train_cat["color"].value_counts(normalize=True)
```

```python
x_train_cat["cut"].value_counts(normalize=True)
```

```python
x_train_cat["clarity"].value_counts(normalize=True)
```

```python
x_train_cat_le = pd.DataFrame(index=x_train_cat.index)
```

```python
cut_label_encoder = {"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
x_train_cat_le["cut"]= x_train_cat["cut"].apply(lambda x:cut_label_encoder[x])
x_train_cat_le.head()
```

```python
x_train_cat_le.cut.unique()
```

```python
x_train_cat_le.cut.nunique()
```

```python
x_train_cat.color.unique()
```

```python
x_train_cat.clarity.unique()
```

```python
color_label_encoder = {"F":1,"E":2,"H":3,"G":4,"D":5,"I":6,"J":7}
x_train_cat_le["color"]= x_train_cat["color"].apply(lambda x:color_label_encoder[x])
x_train_cat_le.head()
```

```python
clarity_label_encoder = {"SI2":1,"VS2":2,"VS1":3,"VVS1":4,"SI1":5,"IF":6,"VVS2":7,"I1":8}
x_train_cat_le["clarity"]= x_train_cat["clarity"].apply(lambda x:clarity_label_encoder[x])
x_train_cat_le.head()
```

# Concatenating the numerical variables and categorical varibales of Train Data

```python
x_train_transformed = pd.concat([x_train_num_rescaled,x_train_cat_le],axis=1)
x_train_transformed.head()
```

```python
# BUilding the model
from sklearn.linear_model import LinearRegression
LR_regressor_object = LinearRegression()
LR_regressor_object.fit(x_train_transformed,y_train)
```

```python
x_test.head()
```

```python
x_test.info()
```

```python
x_test_cat = x_test.select_dtypes(include=["object"])
x_test_cat.columns
```

```python
x_test_num = x_test.select_dtypes(include=["float64"])
x_test_num.columns
```

### Data preprocessing for test data and standard scaling 

```python
x_test_num_rescaled = pd.DataFrame(standard_scaler_object.fit_transform(x_test_num),columns=x_test_num.columns,index=x_test_num.index)
x_test_num_rescaled.head()
```

### Data Preprocessing Categorical Variables

```python
x_test_cat_le = pd.DataFrame(index=x_test_cat.index)
x_test_cat_le.head()
```

```python
x_test_cat_le["cut"]= x_test_cat["cut"].apply(lambda x:cut_label_encoder[x])
x_test_cat_le["color"]= x_test_cat["color"].apply(lambda x:color_label_encoder[x])
x_test_cat_le["clarity"]= x_test_cat["clarity"].apply(lambda x:clarity_label_encoder[x])
x_test_cat_le.head()
```

```python
x_test_transformed = pd.concat([x_test_num_rescaled,x_test_cat_le],axis=1)
```

```python
x_test_transformed.head()
```

## Prediction on test

```python
y_test_pred = LR_regressor_object.predict(x_test_transformed)
tem_df = pd.DataFrame({"Actual values":y_test,"Predicted values":y_test_pred})
tem_df.head()
```

### Mterics

```python
from sklearn import metrics
print("MAE :",metrics.mean_absolute_error(y_test,y_test_pred))
print("MSE :",metrics.mean_squared_error(y_test,y_test_pred))
print("RMSE :",metrics.root_mean_squared_error(y_test,y_test_pred))
```
