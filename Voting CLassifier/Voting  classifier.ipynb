```python
y= df["target"]
x= df.drop(columns=["target"])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lg = LogisticRegression()
gb = GaussianNB()
rf = RandomForestClassifier()
sv = SVC()
```

```python
for clf in (lg,gb,rf,sv):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```

### Using the voting classifier

```python
voting_clf = VotingClassifier(estimators=[('lr',lg),('nb',gb),('rf',rf),('svc',sv)])

for clf in (lg,gb,rf,sv,voting_clf):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```

### Using Hard voting 

```python
voting_clf = VotingClassifier(estimators=[('lr',lg),('nb',gb),('rf',rf),('svc',sv)],voting='hard')

for clf in (lg,gb,rf,sv,voting_clf):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```

### Using Soft Voting

```python
voting_clf = VotingClassifier(estimators=[('lr',lg),('nb',gb),('rf',rf),('svc',sv)],voting='soft')

for clf in (lg,gb,rf,sv,voting_clf):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```
