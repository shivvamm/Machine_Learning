## EDA on dataset 

```python
df.columns
```

```python
x = df[['Area', 'Perimeter', 'MajorAxisLength', 'MinorAxisLength',
       'AspectRation', 'Eccentricity', 'ConvexArea', 'EquivDiameter', 'Extent',
       'Solidity', 'roundness', 'Compactness', 'ShapeFactor1', 'ShapeFactor2',
       'ShapeFactor3', 'ShapeFactor4']]
y =  df[['Class']]
```

```python
x
```

```python
le = LabelEncoder()
y_encoded = pd.DataFrame(le.fit_transform(y),columns=y.columns,index=y.index)
y_encoded = np.ravel(y_encoded)
y_encoded
```

```python
scaler = StandardScaler()
x_rescaled  = pd.DataFrame(scaler.fit_transform(x),columns=x.columns,index=x.index)
x_rescaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x_rescaled,y_encoded, test_size=0.3,random_state=42)
```

## Model building

##  Descision Tree Classifier

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

#### Metrics 

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Random Forest CLassifier

```python
rf = RandomForestClassifier(random_state=0)
rf.fit(x_train,y_train)
```

```python
y_pred =rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Support Vecotr Classifier

#### Liner Kernel

```python
classifier_linear = SVC( C=1.0 ,kernel='linear')
classifier_linear.fit(x_train,y_train)
```

```python
y_pred= classifier_linear.predict(x_test)
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Radial Bias Function Kernel

```python
classifier_RBF = SVC( C=1.0 ,kernel='rbf')
classifier_RBF.fit(x_train,y_train)
```

```python
y_pred= classifier_RBF.predict(x_test)
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Polynoimial Kernel

```python
classifier_poly = SVC( C=1.0 ,kernel='poly')
classifier_poly.fit(x_train,y_train)
```

```python
y_pred= classifier_poly.predict(x_test)
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Sigmoid Kernel

```python
classifier_sigmoid = SVC(C=1.0,kernel='sigmoid')
classifier_sigmoid.fit(x_train,y_train)
```

```python
y_pred= classifier_sigmoid.predict(x_test)
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
