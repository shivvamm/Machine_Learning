# Hold out Validation Approach- Train and Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
result = model.score(x_test,y_test)
print(result)
```

# K-Fold Cross Validation 

```python
from sklearn.model_selection import KFold
model = DecisionTreeClassifier()
k_fold_validation = KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
results = cross_val_score(model,x,y,cv=k_fold_validation)
print(results)
print(np.mean(results))
```

# Stratified K-Fold Cross Validation

```python
x = df.iloc[:,1:]
y =df.iloc[:,0:1]
```

```python
from sklearn.model_selection import StratifiedGroupKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier


skfold = StratifiedGroupKFold(n_splits=2)
model = DecisionTreeClassifier()


print(x.shape, y.shape, df.shape)


try:
    scores = cross_val_score(model, x, y, cv=skfold, groups=y,scoring='accuracy')
    print("Mean cross-validation score:", np.mean(scores))
except Exception as e:
    print("Error during cross-validation:", e)
```

# Leave One Out Cross  Validation

```python
#THis is the most time and resource consuming technique better to go for K-fold and Stratified k-Fold
from sklearn.model_selection import LeaveOneOut
model = DecisionTreeClassifier()
leave_val = LeaveOneOut()
results = cross_val_score(model,x,y,cv=leave_val)
print(results)
```
